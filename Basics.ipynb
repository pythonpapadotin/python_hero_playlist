```python
exam_results = {"Mohit": False, "Nimesh": True, "Nalin": False, "Kriyanshu": True}


for person, result in exam_results.items():
    if result is True:
        print(f"{person} passed the exam")
    else:
        print(f"{person} failed the exam")
```

```python
# Data Types

1. int: Whole Numbers (1, 2, 3, -1, -2, -3), 0?
2. float: Decimal Numbers (2.3, 3.0, -6.7, -5.4) 0.0
3. bool: True, False
4. str: sequence of characters "python", "homeowrk", 'home work', '@#$@#$@ Hisdfjklsdf9900'
5. list: [1, 2, -35, -5.6, 0.0, True, False, 'python', 'home work', ['happy', 'holidays']], changeable, ordered, hetro
6. tuple: (1, 2, -35, -5.6, 0.0, True, False, 'python', 'home work', ['happy', 'holidays']), unchangeable, ordered, hetro
7. dict: {"Mohit": False, "Nimesh": True, "Nalin": False, "Kriyanshu": True}
8. set: {'Mohit', "Rohit", "Mohit"}, unique items
9. None: Represent "no value"
```
